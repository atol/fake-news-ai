# Data Analysis

## Set up

```python
import pandas as pd
import json
import glob
import os

METADATA_FILEPATH = 'dataset/metadata.json'
ARTICLES_FILEPATH = 'dataset/articles'
```

## Load metadata

```python
with open(METADATA_FILEPATH, 'r') as f:
    claims = json.load(f)
```

```python
claims_df = pd.DataFrame(claims)
```

```python
# Print first 10 rows for sanity check
claims_df.head(10)
```

```python
false_claims = claims_df.loc[claims_df['label'] == 0]
false_claims
```

```python
part_true_claims = claims_df.loc[claims_df['label'] == 1]
part_true_claims
```

```python
true_claims = claims_df.loc[claims_df['label'] == 2]
true_claims
```

## Simple statistics

```python
num_claims = len(claims_df.index)
print("Total number of claims:\t\t\t", num_claims)

num_false_claims = len(false_claims.index)
print("\nNumber of false claims:\t\t\t", num_false_claims)

num_part_true_claims = len(part_true_claims.index)
print("Number of partially true claims:\t", num_part_true_claims)

num_true_claims = len(true_claims.index)
print("Number of false claims:\t\t\t", num_true_claims)

percent_false = num_false_claims / num_claims
percent_part_true = num_part_true_claims / num_claims
percent_true = num_true_claims / num_claims

print("\nPercentage of false claims:\t\t", "{0:.2%}".format(percent_false))
print("Percentage of partially true claims:\t", "{0:.2%}".format(percent_part_true))
print("Percentage of true claims:\t\t", "{0:.2%}".format(percent_true))
```

## Load article data

```python
articles = []

for file in glob.glob(os.path.join(ARTICLES_FILEPATH, '*.txt')):
    with open(file) as f:
        body = " ".join(line for line in f)
    
    base = os.path.basename(file)
    file_name = os.path.splitext(base)[0]
    
    article = (os.path.basename(file_name), body)
    articles.append(article)
```

```python
articles_df = pd.DataFrame(articles)
articles_df.columns = ['article_id', 'article']
```

```python
articles_df
```

# Related Articles

```python
related_articles = []

for index, row in claims_df.iterrows():
    for article in row['related_articles']:
        related_articles.append((article, row['id'], row['label']))
```

```python
related_articles_df = pd.DataFrame(related_articles)
related_articles_df.columns = ['article_id', 'claim_id', 'label']
related_articles_df
```

```python
# Associate each article with the labels of the claims that reference it

article_claim_labels = related_articles_df.groupby('article_id')['label'].apply(list).to_dict()
article_claim_labels
```

```python
num_related_articles = related_articles_df.drop_duplicates(subset=['article_id'], keep='first')

print("Number of related articles in claims:", len(num_related_articles))
print("Number of labelled articles:", len(article_claim_labels))
print("Number of total articles:", len(articles_df))
```

```python
# Assign the most frequent label to each article, or the lowest label value in the event of a tie

from statistics import mode

article_single_label = {}

for article_id, label in article_claim_labels.items():
    try:
        article_single_label[article_id] = mode(label)
    except:
        article_single_label[article_id] = min(label)
```

```python
article_single_label
```

```python
labelled_articles_df = articles_df
```

```python
labelled_articles_df['label'] = -1
```

```python
labelled_articles_df
```

```python
for index, row in labelled_articles_df.iterrows():
    key = int(row['article_id'])
    if key in article_single_label:
       labelled_articles_df.at[index, 'label'] = article_single_label[key]
```
